```python
import pandas as pd
import numpy as np
import json
from pprint import pprint
import glob
```

```python
def printTree(nodes, output):
    for node in nodes:
        if node['nodes']:
            printTree(node['nodes'], output)
        else:
            output.append(node)
    return output
```

```python
with open("./maps-master/learn-anything/web-development.json") as file:
    decoded = json.load(file)
    teste = printTree(decoded['nodes'],[])
    print(len(teste))
    pprint(teste)
```

```python
json_files = glob.glob('./maps-master/learn-anything/**/*.json', recursive=True)
final_output_array = []
for json_file in json_files:
    with open(json_file) as file:
        json_file_decoded = json.load(file)
        output_array = printTree(json_file_decoded['nodes'],[])
        final_output_array += output_array
```

```python
df = pd.DataFrame(final_output_array)
```

```python
df.to_csv("learn-anything_urls.csv")
```
